```python
#Find the largest and Smallest Number
a=int(input("Enter the value of first Number"))
b=int(input("Enter the value of second Number"))
c=int(input("Enter the value of third Number"))
#For largest
if(a>b and a>c):
  print(a,"is Largest")
elif(b>a and b>c):
  print(b,"is Largest")
else:
  print(c,"is Largest")

# For smallest
if(a<b and a<c):
  print(a,"is Smallest")
elif(b<a and b<c):
  print(b,"is Smallest")
else:
  print(c,"is Smallest")
```
