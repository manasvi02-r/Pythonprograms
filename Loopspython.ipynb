```python
#Palindrome
a=int(input("Enter the number:"))
temp=a # Stores the original number for later comparison.
rev=0  # Initializes a variable to store the reversed number.
while a>1:

    # Extracts the last digit of the number
    digit=a%10
    # Adds the extracted digit to the reversed number in the correct place value.
    rev=(rev*10)+digit
    # Removes the last digit from the original number.
    a=a//10

# Prints the final reversed number.
print(rev)

#comparing original value and reverse value for palindrome number
if(temp==rev):
    print("Num is Palindrome")
else:
    print("Num is not Palindrome")
```

```python
#Armstrong
a=int(input("Enter the number:"))
temp=a
sum=0
while a>0:

    # Extract the last digit of 'a' using the modulo operator.
    digit=a%10
    # Add the cube of the extracted digit to the running total 'sum'.
    sum = sum + digit * digit * digit
    # Remove the last digit from 'a' using integer division, preparing for the next iteration.
    a=a//10

# After the loop finishes, check if the calculated 'sum' is equal to the original number stored in 'temp'
if(sum==temp):
    print("Num is Armstrong")
else:
    print("Num is not Armstrong")
```
